```python
import torch as th
import numpy as np
import matplotlib.pyplot as plt
import src.potentials as pot
import src.algorithms as algo
import src.utils as util
import os
import pandas as pd
from pathlib import Path

device = th.device('cuda' if th.cuda.is_available() else 'cpu')


N = 5000   # number of samples
check_iter = 10
show_plot = False
save_sample = 1000
x_init = th.zeros([N,1]).to(device)-1

methods = {
            'ULA': True,
            'dilation':True,
            'tempering':True,
            'diffusion':True,
            'DAZ':True
}


means = th.tensor([[0.,0],
                   [1,0],
                   [0,1],
                   [1,1]]).to(device)

sigmas = th.tensor([[0.2,0.2],
                   [0.1,0.2],
                   [0.3,0.1],
                   [.1,.1]]).to(device)

L=1/sigmas.min()**2
a = 1/sigmas.max()**2
step = a/L**2
times = th.Tensor(np.arange(0,100,step.cpu()))
T=1e1
taus = th.exp(-times/T)*.99
print(f'step: {step}')
mixture_weights = th.Tensor([0.2,0.4,0.2,.2]).to(device)
gmm = pot.GMM_diffusion(means = means, sigmas=sigmas, mixture_weights=mixture_weights)

folder = f'results/gmm_2d/T_{T}'
Path(folder).mkdir(exist_ok=True)
# ground truth samples
sample = gmm.sample(1000,tau=0.0)
th.save(sample,f'{folder}/gt_sample')


direct_sample_hist, direct_sample_bins = th.histogramdd(sample.cpu(),bins=200,density=True)
x_vals = (direct_sample_bins[0][1:]+direct_sample_bins[0][:-1])/2
y_vals = (direct_sample_bins[1][1:]+direct_sample_bins[1][:-1])/2
xmin = x_vals.min()
xmax = x_vals.max()
ymin = y_vals.min()
ymax = y_vals.max()
plt.figure()
plt.imshow(
        direct_sample_hist,
        origin="lower",
        extent=(xmin, xmax, ymin, ymax),cmap='hot')


xx,yy = th.meshgrid(x_vals,y_vals)
coordinates = th.cat([xx[...,None],yy[...,None]],dim=-1)
coordinates = coordinates.view((coordinates.shape[:-1].numel(),coordinates.shape[-1])).to(device)
gt_hist = gmm(coordinates,tau=0).view(xx.shape).cpu().numpy()
plt.figure()
plt.imshow(
        gt_hist,
        origin="lower",
        extent=(xmin, xmax, ymin, ymax),cmap='hot')
```
